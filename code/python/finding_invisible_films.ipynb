```python
def create_prompt(row: pd.Series) -> str:
    """
    Convert a DataFrame row into a text prompt for LLM rating prediction,
    dynamically including all available non-null fields.
    """
    prompt = "Based on the following film information, predict the rating (0-10):\n\n"
    
    # Define columns to exclude
    excluded_columns = ['vote_average', 'vote_count', 'video', 'backdrop_path', 'poster_path', 'id', 'imdb_id']
    
    # Iterate through all columns
    for column, value in row.items():
        # Skip if column is in excluded list or value is null
        if column in excluded_columns or pd.isna(value):
            continue
            
        # Skip empty lists or empty strings
        if (isinstance(value, list) and len(value) == 0) or \
           (isinstance(value, str) and value.strip() == ""):
            continue
            
        # Check for more complex cases like string representations of empty lists
        if isinstance(value, str) and value.strip() in ["[]", "{}"]:
            continue
            
        # Capitalize column name for readability and include the value
        prompt += f"{column.capitalize()}: {value}\n"
    
    prompt += "\nPredicted Rating (0-10): "
    
    return prompt
```

```python
print(create_prompt(df.sample(n=1).iloc[0]))
```
